```python
# ============================================
# LESSON 1: YOUR FIRST AI TEXT GENERATOR
# ============================================
# Goal: Generate text with just 3 lines of code!

# Step 1: Install (run this in your terminal ONCE)
# pip install transformers

# Step 2: Import what we need
from transformers import pipeline

# Step 3: Create your AI text generator
# This downloads a small AI model (might take 1-2 minutes first time)
generator = pipeline('text-generation', model='gpt2')

# Step 4: Use it!
result = generator("In the year 2050")

# Step 5: See the result
print(result[0]['generated_text'])

# ============================================
# TRY THESE:
# ============================================
# Change "Once upon a time" to:
# - "The secret to happiness is"
# - "In the year 2050"
# - "My favorite food is"

# That's it! You just used an AI language model! ðŸŽ‰
```
